```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from bs4 import BeautifulSoup
from tqdm import tqdm

# Setup driver
driver = webdriver.Chrome()

def select_from_dropdown(container_id, option_text):
    # Click the fake dropdown
    driver.find_element(By.ID, container_id).click()
    time.sleep(1)
    # Click the actual option
    driver.find_element(By.XPATH, f"//li[normalize-space(text())='{option_text}']").click()
    time.sleep(1)

for year in tqdm([2023, 2024]):
    for rnd in range(1, 6):
        ##### Part 1
        driver.get("https://josaa.admissions.nic.in/applicant/seatmatrix/openingclosingrankarchieve.aspx")
        time.sleep(5)

        ##### Part 2
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlYear_chosen', year)
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlroundno_chosen', rnd)
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlInstype_chosen', 'ALL')
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlInstitute_chosen', 'ALL')
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlBranch_chosen', 'ALL')
        select_from_dropdown('ctl00_ContentPlaceHolder1_ddlSeatType_chosen', 'ALL')
        driver.find_element(By.ID, 'ctl00_ContentPlaceHolder1_btnSubmit').click()
        
        time.sleep(100)
        
        ##### Part 3
        driver.page_source
        time.sleep(10)

        ##### Part 4
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        ##### Part 5
        headers = []
        for sp in soup.find_all('th'):
            headers.append(sp.text.replace('\n','').strip())
        headers.append('rnd')

        ##### Part 6
        Data = []
        rows = soup.find_all('tr')[1:-1]
        for row in tqdm(rows):  ### 99 Out of 11687 data rows
            time.sleep(0.05)
            data = []
            data.append(row.find_all('td')[0].text.replace('\n','').strip())
            for j in range(0,6):
                data.append(row.find_all('span')[j].text.strip())
            data.append(rnd)
            Data.append(data)

        ##### Part 7
        df = pd.DataFrame(Data , columns = headers)
        df.to_csv(f"OR_CR_{year}_{rnd}.csv",index = False)
    print(f"Scraped {year} year successfully")
    driver.quit()
```
