```python
import pandas as pd
import glob

# Get all files matching the pattern OR_CR*.csv
file_list = glob.glob("OR_CR*.csv")

# Read and concatenate all files into one DataFrame
merged_df = pd.concat((pd.read_csv(file) for file in file_list), ignore_index=True)

# Show the shape or head of merged DataFrame
print(f"Total files merged: {len(file_list)}")
print("Combined shape:", merged_df.shape)
merged_df
```

```python
merged_df.dtypes
```

```python
import re

# Apply regex to reduce multiple spaces to a single space
merged_df['Institute'] = merged_df['Institute'].astype(str).apply(lambda x: re.sub(r'\s+', ' ', x.strip()))
```

```python
def classify_institute(name):
    name = str(name).lower()
    name = name.replace('\xa0', ' ')  # remove non-breaking spaces

    if 'indian institute of technology' in name:
        return 'Indian Institute of Technology'
    elif 'national institute of technology' in name:
        return 'National Institute of Technology'
    elif 'indian institute of information technology' in name:
        return 'Indian Institute of Information Technology'
    else:
        return 'Governement Funded Technical Institute'
        
merged_df['Institute Type'] = merged_df['Institute'].apply(classify_institute)
merged_df
```

```python
# See how many unique program names exist
unique_institutes = merged_df['Institute'].unique()

# Sort and print them
unique_institutes = sorted(unique_institutes)
print(f"Total unique programs: {len(unique_institutes)}")
for inst in unique_institutes:
    print(inst)
```

```python
# See how many unique program names exist
unique_programs = merged_df['Academic Program Name'].unique()

# Sort and print them
unique_programs = sorted(unique_programs)
print(f"Total unique programs: {len(unique_programs)}")
for prog in unique_programs:
    print(prog)
```

```python
import re

def split_program(program):
    # Ensure string and strip spaces
    program = str(program).strip()
    
    # Extract content inside parentheses that contains any digit (for degree)
    match = re.search(r'\(([^)]*\d[^)]*)\)', program)
    
    if match:
        degree = match.group(1).strip()
        branch = program.replace(f"({degree})", "").strip()
    else:
        degree = None
        branch = program.strip()
    
    return pd.Series([branch, degree])

# Apply to the column
merged_df[['Branch', 'Degree']] = merged_df['Academic Program Name'].apply(split_program)
```

```python
# Define mapping dictionary
quota_mapping = {
    'AI': 'All India Quota',
    'HS': 'Home State Quota',
    'OS': 'Other State Quota',
    'GO': 'Gender-Neutral Open Category',
    'JK': 'Jammu & Kashmir Quota',
    'LA': 'Ladakh Domicile Quota'
}

# Replace in Quota column (if present)
if 'Quota' in merged_df.columns:
    merged_df['Quota'] = merged_df['Quota'].replace(quota_mapping)
```
