```python
import pandas as pd

# Get common columns
common_cols = df1.columns

# Extract only the common columns from df1
df2_common = df2[common_cols]

# Create a new DataFrame by interleaving rows
interleaved_rows = []
for i in range(len(df2)):
    interleaved_rows.append(df2_common.iloc[i])
    interleaved_rows.append(df1.iloc[i])

# Convert to DataFrame
merged_df = pd.DataFrame(interleaved_rows)

# Optionally reset index
merged_df.reset_index(drop=True, inplace=True)
```
